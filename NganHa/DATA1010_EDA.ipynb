```python
# Distribution of Flight Prices

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.histplot(df['price'], bins=30, kde=True, color='skyblue')
plt.title("Distribution of Flight Prices")
plt.xlabel("Price")
plt.ylabel("Number of Flights")
plt.show()
```

```python
# Price vs. Total Duration

plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='total_duration', y='price', hue='travel_class', alpha=0.7)
plt.title("Flight Price vs Total Duration")
plt.xlabel("Total Duration (minutes)")
plt.ylabel("Price")
plt.grid(True)
plt.show()
```

```python
# Price vs. Carbon Emissions

plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='carbon_this', y='price', hue='travel_class', style='is_direct', alpha=0.7)
plt.title("Flight Price vs Carbon Emissions")
plt.xlabel("Carbon Emissions (g CO₂)")
plt.ylabel("Price")
plt.grid(True)
plt.show()
```

```python
# Price per Minute vs Number of Stops

plt.figure(figsize=(4, 6))
sns.scatterplot(data=df, x='n_stops', y='price_per_minute', hue='travel_class', alpha=0.6)
plt.title("Price per Minute vs Number of Stops")
plt.xlabel("Number of Stops")
plt.ylabel("Price per Minute")
plt.grid(True)
plt.show()
```

```python
# Price vs. Days to Departure

plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='days_to_departure', y='price', hue='delayed_any', palette={True: 'red', False: 'green'})
plt.title("Flight Price vs Days to Departure")
plt.xlabel("Days to Departure")
plt.ylabel("Price")
plt.grid(True)
plt.show()
```

```python
# Price Distribution by Number of Stops

import seaborn as sns

sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.boxplot(x='n_stops', y='price', data=df)
plt.title('Price Distribution by Number of Stops')
plt.xlabel('Number of Stops')
plt.ylabel('Price')
plt.tight_layout()
plt.show()
```

```python
# Carbon Difference vs. Price

plt.figure(figsize=(10, 6))
sns.scatterplot(x='carbon_diff_pct', y='price', hue='n_stops', palette='viridis', data=df)
plt.title('Carbon Difference (%) vs. Price')
plt.xlabel('Carbon Difference Percentage')
plt.ylabel('Price')
plt.axvline(0, color='red', linestyle='--', label='Baseline')
plt.legend(title='Number of Stops')
plt.tight_layout()
plt.show()
```

```python
# Delay Rate by Number of Stops

plt.figure(figsize=(8, 5))
delay_rates = df.groupby('n_stops')['delayed_any'].mean().reset_index()
sns.barplot(x='n_stops', y='delayed_any', data=delay_rates)
plt.title('Delay Rate by Number of Stops')
plt.xlabel('Number of Stops')
plt.ylabel('Delay Rate')
plt.ylim(0, 1)
plt.tight_layout()
plt.show()
```
